## Import required Python libraries

```python
import numpy as np
from netCDF4 import Dataset
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import pandas as pd
```

## Read in NetCDF4 file

Use the netCDF4 package to read the data and then print to show information about the data, including the variables of interest.

```python
data = Dataset('exampleMERRA2data.nc', mode='r')
print(data)
```

## Get list of coordinates

Approximate rectangle of Greenland coordinates: 

Zonal: 80 degrees W to 10 degrees W (lon: -80 to -10)

Meridional: 55 degrees N to 85 degrees N (lat: +55 to +85)

```python
lons = data.variables['lon'][:]
lats = data.variables['lat'][:]
```

## Plot corners to check coordinates

```python
x = np.array([lons[160],lons[272],lons[160],lons[272]])
y = np.array([lats[290],lats[290],lats[350],lats[350]])

fig = plt.figure(figsize=(8,4))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_global()
ax.coastlines()
ax.scatter(x,y,color='r',transform=ccrs.PlateCarree())
plt.show()
```
